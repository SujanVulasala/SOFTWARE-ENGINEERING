```python
# -*- coding: utf-8 -*-
"""SE Ver 4.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1EV1Zfp70fl8Xk6FzIFF46XuXLX7WGFrr#scrollTo=hYk2c2Ns5P_Z
"""

import matplotlib.pyplot as plt

x_values = []
y_values = []

with open('/content/data2.txt', 'r') as file:
    while True:
        try:
            line1 = file.readline().strip()
            line2 = file.readline().strip()
            line3 = file.readline().strip()
            line4 = file.readline().strip()

            if not line1:  # Check if we've reached the end of the file
                break

            a = float(line1.split('=')[1])
            b = float(line2.split('=')[1])
            c = float(line3.split('=')[1])
            x = float(line4.split('=')[1])

            y = a * x * x + b * x + c

            x_values.append(x)
            y_values.append(y)

        except (IndexError, ValueError):
            print("Error processing lines. Please check the file format.")
            break

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('Graph of y = ax^2 + bx + c')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
